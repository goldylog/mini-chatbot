```python
from langchain.document_loaders import TextLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.schema import Document
import os
```

```python
file_paths = [
    "/content/drive/MyDrive/dongduck/25/컴네/data/taskdata/컴퓨터네트워크_1주차.txt",
    "/content/drive/MyDrive/dongduck/25/컴네/data/taskdata/컴퓨터네트워크_2주차.txt",
    "/content/drive/MyDrive/dongduck/25/컴네/data/taskdata/컴퓨터네트워크_3주차.txt",
    "/content/drive/MyDrive/dongduck/25/컴네/data/taskdata/컴퓨터네트워크_4주차.txt",
    "/content/drive/MyDrive/dongduck/25/컴네/data/taskdata/컴퓨터네트워크_5주차.txt",
    "/content/drive/MyDrive/dongduck/25/컴네/data/taskdata/컴퓨터네트워크_6주차.txt",
    "/content/drive/MyDrive/dongduck/25/컴네/data/taskdata/컴퓨터네트워크중간고사예상문제.txt"
]
```

```python
# 2. 문서 로드
all_documents = []
for path in file_paths:
    try:
        with open(path, "r", encoding="utf-8-sig") as f:
            content = f.read()
    except UnicodeDecodeError:
        with open(path, "r", encoding="cp949") as f:  # 대체 인코딩
            content = f.read()
    all_documents.append(Document(page_content=content))
```

```python
# 3. 문서 분할 함수
def split_docs(documents, chunk_size=1000, chunk_overlap=100):
    text_splitter = RecursiveCharacterTextSplitter(chunk_size=chunk_size, chunk_overlap=chunk_overlap)
    return text_splitter.split_documents(documents)

# 4. 분할된 문서
docs = split_docs(all_documents)
docs[:1]
```

```python
#OpenAI의 임베딩 모델 사용
from langchain_openai import OpenAIEmbeddings
api_key="YOUR_API_KEY"
embeddings = OpenAIEmbeddings(model="text-embedding-ada-002", api_key=api_key)
```

```python
# Chromdb에 벡터 저장
from langchain.vectorstores import Chroma
db = Chroma.from_documents(docs, embeddings, persist_directory=".")
```

```python
from langchain.chat_models import ChatOpenAI
model_name = "gpt-3.5-turbo"  #GPT-3.5-turbo 모델 사용
llm = ChatOpenAI(model_name=model_name, api_key=api_key)

# Q&A 체인을 사용하여 쿼리에 대한 답변 얻기
from langchain.chains.question_answering import load_qa_chain
qa_chain = load_qa_chain(llm, chain_type="stuff",verbose=True)

# 쿼리를 작성하고 유사성 검색을 수행하여 답변을 생성,따라서 txt에 있는 내용을 질의해야 합니다
query = "modem이란?"
matching_docs = db.similarity_search(query)
answer =  qa_chain.run(input_documents=matching_docs, question=query)
answer
```

```python
from langchain.memory import ConversationBufferMemory
from langchain.chains import ConversationalRetrievalChain

# memory 객체 생성
memory = ConversationBufferMemory(memory_key="chat_history", return_messages=True)

# 대화형 QA 체인 생성
qa_chain = ConversationalRetrievalChain.from_llm(
    llm = ChatOpenAI(
    temperature=0,
    model_name="gpt-3.5-turbo",  # 또는 gpt-4
    openai_api_key="YourAPIkey"      # 실제 본인의 OpenAI 키 입력
),
    retriever=db.as_retriever(),  # 기존 Chroma DB를 그대로 사용
    memory=memory
)
```

```python
def chatbot_interface(user_input, chat_history):
    if not user_input:
        return chat_history, chat_history

    # LangChain chain 실행
    result = qa_chain({"question": user_input})
    answer = result["answer"]

    # Gradio 대화기록 업데이트
    chat_history.append((user_input, answer))
    return chat_history, chat_history
```

```python
# Q&A 함수 정의
def answer_question(query):
    matching_docs = db.similarity_search(query)
    answer = qa_chain.run(input_documents=matching_docs, question=query)
    return answer
```

```python
import gradio as gr
```

```python
# 초기화 함수
def reset_memory():
    memory.clear()
    return [], []
```

```python
# Gradio Blocks 기반 UI 구성
with gr.Blocks() as demo:
    chatbot = gr.Chatbot()
    msg = gr.Textbox(label="질문을 입력하세요")
    state = gr.State([])  # Gradio 내부에서 대화 상태 저장용

    # 사용자 입력 처리 함수 (LangChain QA chain 활용)
    def chatbot_interface(user_input, chat_history):
        if not user_input:
            return chat_history, chat_history
        result = qa_chain({"question": user_input})
        answer = result["answer"]
        chat_history.append((user_input, answer))
        return chat_history, chat_history

    # 제출 시 응답
    msg.submit(fn=chatbot_interface, inputs=[msg, state], outputs=[chatbot, state])

    # 초기화 버튼
    reset_btn = gr.Button("초기화")
    reset_btn.click(fn=reset_memory, inputs=[], outputs=[chatbot, state])

demo.launch(debug=True)



# Gradio UI 구성
#demo = gr.Interface(
 #   fn=answer_question,
  #  inputs=gr.Textbox(lines=2, label="질문을 입력하세요"),
   # outputs=gr.Textbox(label="답변"),
    #title=" LangChain 기반 AI Q&A",
    #description="업로드된 문서를 기반으로 GPT-3.5가 답변을 생성합니다."
#)
#with gr.Blocks() as demo:
#    chatbot = gr.Chatbot()
#    msg = gr.Textbox(label="질문을 입력하세요")
#    state = gr.State([])  # chat_history를 기억하는 상태 변수

    # 초기화 버튼
#    reset_btn = gr.Button("초기화")
#    reset_btn.click(fn=reset_memory, inputs=[], outputs=[chatbot, state])

#    msg.submit(fn=chatbot_interface, inputs=[msg, state], outputs=[chatbot, state])

#demo.launch(debug=True)
```
